```python
import pandas as pd
import numpy as np

df_sell = pd.read_csv('MS-b1-sell.csv')
df_sell
```

```python
df_inventory = pd.read_csv('MS-b1-inventory.csv')
df_inventory
```

```python
res_df = pd.DataFrame(columns=['date', 'apple', 'pen'])
prev_date = '2006-01-01'
total_apples = 0
total_pens = 0

for row in df_sell.values:
    cur_date = row[0]
    product = str(row[1])[6:8]
    if cur_date != prev_date:
        prev_date = cur_date
        cur_row = pd.DataFrame(columns=['date', 'apple', 'pen'], data =[[cur_date, total_apples, total_pens]])
        res_df = pd.concat([res_df, cur_row], ignore_index = True)
        total_apples = total_pens = 0
    if product == 'ap':
        total_apples += 1
    else:
        total_pens += 1
res_df



        
```
